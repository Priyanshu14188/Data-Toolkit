Q2. How does broadcasting work in NumPy?

Answer:

Broadcasting in NumPy means performing operations (like addition, subtraction, multiplication, etc.) on arrays of different shapes without explicitly copying data.

Broadcasting Rules

Compare shapes of both arrays starting from the rightmost dimension.

Two dimensions are compatible if:

They are equal, OR

One of them is 1.

If dimensions are incompatible → ValueError.

Example 1: Adding a scalar

        import numpy as np

        a = np.array([1, 2, 3])
        b = 5

        print(a + b)   # Broadcasting scalar


        Output: [6 7 8]
        (NumPy "stretches" 5 to [5,5,5])

Example 2: Different shapes

      a = np.array([[1, 2, 3],
                    [4, 5, 6]])   # shape (2,3)

      b = np.array([10, 20, 30])  # shape (3,)

      print(a + b)


      Output:

      [[11 22 33]
      [14 25 36]]


Here, b is broadcasted to shape (2,3).


Example 3: Incompatible shapes

    a = np.array([[1, 2, 3],
                  [4, 5, 6]])   # shape (2,3)

    b = np.array([10, 20])      # shape (2,)

    print(a + b)


Error: ValueError: operands could not be broadcast together
(Because (2,3) and (2,) are not compatible)

Q4. Explain the use of the groupby() method in Pandas?

Answer:

In Pandas, the groupby() method is used to split data into groups, apply some function/operation on each group, and then combine the results.

This is often called the Split–Apply–Combine strategy.

Steps:

* Split → Divide the DataFrame into groups based on column(s).

* Apply → Perform an operation (like sum, mean, count, etc.) on each group.

* Combine → Put the results back together into a new DataFrame/Series.

Q6. What are the differences between NumPy arrays and Python lists?

Answer:

NumPy Arrays:

* Homogeneous → all elements must be of the same type (e.g., all int or all float).

* Fast & efficient → uses contiguous memory and C-optimized code.

* Supports vectorized operations → you can add, multiply, or apply functions to entire arrays without loops.

* Fixed size → resizing creates a new array.

* Rich mathematical, statistical, and linear algebra functions (e.g., mean, sum, dot).

* Handles multi-dimensional arrays (2D, 3D, tensors).

Python Lists:

* Heterogeneous → can store different data types together (int, str, float, etc.).

* Slower → extra memory overhead for references and no vectorization.

* No direct mathematical operations → need loops or list comprehensions.

* Dynamic size → can grow or shrink easily.

* Limited built-in functions (no advanced math like NumPy).

* Only 1D structure → 2D/3D requires lists of lists (inefficient).

Q8. What does the term “vectorized operation” mean in NumPy?

Answer:

Vectorized Operation in NumPy

A vectorized operation means applying an operation directly on entire arrays (or vectors) instead of looping through elements one by one.

NumPy is optimized in C under the hood, so vectorized operations are much faster than writing Python loops.


        # Without vactorized operation
        import numpy as np

        arr = np.array([1, 2, 3, 4, 5])
        result = []
        for x in arr:
            result.append(x * 2)
        print(result)

        #Output
        [2, 4, 6, 8, 10]


        # Vactorized operation
        import numpy as np

        arr = np.array([1, 2, 3, 4, 5])
        result = arr * 2
        print(result)

Q13. Why is handling missing data important in Pandas?

Q15. How does NumPy handle multidimensional arrays?

Answer:

NumPy handles multidimensional arrays using its special data structure called the ndarray (N-dimensional array).

* Here’s how it works:

    * Shape → Defines the size along each dimension (e.g., (3, 4) means 3 rows × 4 columns).

    * Axes (dimensions) → A 1D array has 1 axis, a 2D array has 2 axes (rows & columns), a 3D array has 3 axes, etc.

    * Indexing → You can access elements using row/column indices (e.g., arr[1, 2] → element at row 1, column 2).

    * Efficient storage → All elements are stored in a single contiguous memory block, making it much faster than nested Python lists.

    * Operations → You can perform element-wise operations, matrix multiplication, slicing, reshaping, and broadcasting across dimensions.

Q17. Explain the difference between apply() and map() in Pandas?

Answer:

1. map()

* Works element-wise on a Series (not DataFrame).

* Can take a function, dictionary, or Series as an argument.

* Best for applying a simple transformation on one column (Series).

        import pandas as pd

        s = pd.Series([1, 2, 3, 4])
        print(s.map(lambda x: x**2))

        Output
        0     1
        1     4
        2     9
        3    16
        dtype: int64

2. apply()

* Works on both DataFrame and Series.

* Applies a function along an axis (row-wise or column-wise).

* More flexible and powerful than map().

* Suitable for complex row/column operations.

        df = pd.DataFrame({
            "A": [1, 2, 3],
            "B": [10, 20, 30]
        })
        print(df.apply(lambda x: x.sum(), axis=0))
        print(df.apply(lambda x: x.sum(), axis=1))   

        Output
        A     6
        B    60
        dtype: int64

        0    11
        1    22
        2    33
        dtype: int64

Q21.  Why is NumPy’s array slicing faster than Python’s list slicing?

Answer:

NumPy’s array slicing is faster than Python’s list slicing:

* Contiguous Memory – NumPy arrays are stored in continuous memory blocks, so slicing just creates a view (no data copy). Lists are scattered objects in memory, requiring more work.

* Views, Not Copies – When you slice a NumPy array, it doesn’t copy data, it just creates a view into the original array. Python lists create a new list when sliced.

* Low-Level C Implementation – NumPy is written in C and optimized for speed. List slicing uses Python’s interpreted loops, which are slower.

* Fixed Data Types – NumPy arrays use a single dtype for all elements, making slicing operations straightforward. Python lists can hold mixed types, so slicing has extra overhead.
